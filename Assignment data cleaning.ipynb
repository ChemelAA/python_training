```python
# droping "unnamed: 0" and "unnamed: 8" columns

del data['Unnamed: 0']
del data['Unnamed: 8']
data
```

```python
# Changing weird column names

new_columns = [
    'imbd_id', 
    'title', 
    'release_year', 
    'genre', 
    'duration', 
    'country', 
    'content_rating', 
    'director', 
    'income', 
    'votes', 
    'score'
]

data.columns = new_columns
data
```

```python
# Remove empty row

data.dropna(how='all', inplace=True)
data
```

```python
# Remove symbols besides English characters, numbers and punctuation

import re
columns_to_clean = ['title', 'genre', 'director']
for col in columns_to_clean:
    data[col] = data[col].apply(lambda text: re.sub(r'[^\x00-\x7F]+', '', text))
data
```

```python
# Clean date column

data['new_date'] = pd.to_datetime(data['release_year'], errors='coerce')
data
```

```python
data[data.new_date.isna()]
```

```python
data.loc[70, 'release_year'] = '1951-03-06'
data['new_date'] = pd.to_datetime(data['release_year'], errors='coerce')
data[data.new_date.isna()]
```

```python
# This two row have mistakes in dates, so we can remove it
data.dropna(subset=['new_date'], inplace=True)
data
```

```python
# Lets change the date format
data['release_date'] = data['new_date'].dt.strftime('%d/%m/%Y')
del data['release_year']
del data['new_date']
data
```

```python
# Duration
import numpy as np
data['duration'] = data['duration'].str.replace(r'\D', '', regex=True)
data['duration'] = data['duration'].replace('', np.nan)
data['duration'] = data['duration'].fillna(-1).astype(int) # Filling absent durations as -1
data
```

```python
# Country

data['country'] = data['country'].str.replace(r'[^a-zA-Z ]', '', regex=True)
replacements = {'New Zeland': 'New Zealand', 'New Zesland': 'New Zealand', 'US': 'USA', 'US.': 'USA'}
data['country'] = data['country'].replace(replacements)
data
```

```python
# Rating

replacements_rating = {'Approved': '', 'Not Rated': '', 'Unrated': '', np.nan: ''}
data['content_rating'] = data['content_rating'].replace(replacements_rating)
data
```

```python
# Votes

data['votes'] = data['votes'].str.replace(r'\D', '', regex=True)
data['votes'] = data['votes'].replace('', np.nan)
data['votes'] = data['votes'].fillna(-1).astype(int)
data
```

```python
# Income

data['income'] = data['income'].replace({'o': '0'})
data['income'] = data['income'].str.replace(r'\D', '', regex=True)
data['income'] = data['income'].replace('', np.nan)
data['income'] = data['income'].fillna(-1).astype('int64')
data
```

```python
# Score

data['score'] = (data['score']
                 .str.replace(r'[,:]', '.', regex=True)
                 .str.lstrip('0')
                 .str.rstrip('.')
                 .str.replace(r'[^\d.]', '', regex=True)
                 .str.replace(r'\.{2,}', '.', regex=True)
                 .astype(float)
                )
data
```
